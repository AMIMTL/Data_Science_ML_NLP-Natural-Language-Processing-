## 2. Text Preprocessing with spaCy

In addition to the lowercasing and special character removal from the previous assignment, within the cleaned Description column:
* Tokenize the text
* Lemmatize the text
* Remove stop words

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Tokenize the text
def tokenization(text:str):
  doc = nlp(text)
  output = [token.text for token in doc]
  return output
```

```python
df_children_books["Description_Clean"] =df_children_books["Description_Clean"].apply(tokenization)
df_children_books["Description_Clean"]
```

```python
# Lemmatize the text
def lemmatization(text: list):
  # Process the list of tokens with nlp to get Doc objects
  doc = nlp(" ".join(text))
  output = [token.lemma_ for token in doc]
  return output
```

```python
df_children_books["Description_Clean"] = df_children_books["Description_Clean"].apply(lemmatization)
df_children_books["Description_Clean"]
```

```python
# Remove stop words
print(nlp.Defaults.stop_words, "\n" , type(nlp.Defaults.stop_words))
print(nlp.Defaults.prefixes , "\n" ,  type(nlp.Defaults.prefixes))
```

```python
def remove_stopwords(tokens: list) -> list:
  # Process the list of tokens with nlp to get Doc objects
  doc = nlp(" ".join(tokens))
  output = [token.lemma_ for token in doc if not token.is_stop]
  return output
```

```python
df_children_books["Description_Clean"] = df_children_books["Description_Clean"].apply(remove_stopwords)
df_children_books["Description_Clean"]
```

## 3. Count Vectorizer

1. Vectorize the cleaned and normalized text using Count Vectorizer with the default parameters
2. Modify the Count Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
3. Use the updated Count Vectorizer to identify the:
* Top 10 most common terms
* Top 10 least common terms that appear in at least 10% of the documents
4. Create a horizontal bar chart of the top 10 most common terms

----------------------------------------------------

1. Vectorize the cleaned and normalized text using Count Vectorizer with the default parameters

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
dtm = cv.fit_transform(df_children_books["Description_Clean"].apply(lambda X: " ".join(X)))
dtm_df = pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out())
dtm_df
```

2. Modify the Count Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%

```python
cv2 = CountVectorizer(stop_words='english', min_df= 0.10)
dtm2 = cv2.fit_transform(df_children_books.Description_Clean.apply(lambda X: " ".join(X)))
dtm2_df = pd.DataFrame(dtm2.toarray(), columns = cv2.get_feature_names_out())
dtm2_df
```

3. Use the updated Count Vectorizer to identify the:
* Top 10 most common terms
* Top 10 least common terms that appear in at least 10% of the documents

```python
#Top 10 most common terms
dtm2_df.sum().nlargest(10)
# or dtm2_df.sum().sort_values(ascending=False).head(10)
```

```python
#Top 10 least common terms that appear in at least 10% of the documents
dtm2_df.sum().nsmallest(10)
# or dtm2_df.sum().sort_values(ascending=True).head(10)
```

4. Create a horizontal bar chart of the top 10 most common terms

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
dtm2_df.sum().nlargest(10).sort_values().plot(kind ='barh', color='cyan')
plt.show()
```

## 4. TF-IDF Vectorizer

1. Vectorize the cleaned and normalized text using TF-IDF Vectorizer with the default parameters
2. Modify the TF-IDF Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
* Set a maximum document frequency of 50%
3. Using the updated TF-IDF Vectorizer, create a  horizontal bar chart of the top 10 most highly weighted terms
4. Compare the Count Vectorizer bar chart from the previous assignment with the TF-IDF Vectorizer bar chart and note the differences in the top term lists

----------------------------------------------------------------------------

1. Vectorize the cleaned and normalized text using TF-IDF Vectorizer with the default parameters

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer()
tfidf = tfv.fit_transform(df_children_books.Description_Clean.apply(lambda X: " ".join(X)))
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfv.get_feature_names_out())
tfidf_df
```

2. Modify the TF-IDF Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
* Set a maximum document frequency of 50%

```python
tfv2 = TfidfVectorizer(stop_words='english', min_df=0.1 , max_df=0.5)
tfidf2 = tfv2.fit_transform(df_children_books.Description_Clean.apply(lambda X: " ".join(X)))
tfidf2_df = pd.DataFrame(tfidf2.toarray(), columns = tfv2.get_feature_names_out())
tfidf2_df
```

3. Using the updated TF-IDF Vectorizer, create a horizontal bar chart of the top 10 most highly weighted terms

```python
tfidf2_df.sum().nlargest(10)
```

```python
tfidf2_df.sum().nlargest(10).sort_values(ascending=1).plot(kind= 'barh', color ='deeppink')
plt.show()
```

4. Compare the Count Vectorizer bar chart from the previous assignment with the TF-IDF Vectorizer bar chart and note the differences in the top term lists

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

dtm2_df.sum().nlargest(10).sort_values().plot(kind ='barh', color='cyan', ax=ax[0])
ax[0].set_title('Count Vectorizer Top 10 Terms')

tfidf2_df.sum().nlargest(10).sort_values().plot(kind= 'barh', color ='deeppink', ax=ax[1])
ax[1].set_title('TF-IDF Vectorizer Top 10 Terms')

plt.tight_layout()
plt.show()
```
